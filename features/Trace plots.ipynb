```python
def plot_cell(file):
    abf = pyabf.ABF(file)
    print(file)
    plt.figure(figsize=(10,10))
    for i in range(12):
        abf.setSweep(i)
        plt.subplot(6,2,i + 1)
        plt.title(str(i))
        plt.plot(abf.sweepX, abf.sweepY, lw=.5)
    plt.show()
```

```python
def plot_abf(name):
    abf = pyabf.ABF(name)
    plt.figure(figsize=(8, 5))
    for sweepNumber in range(20):
        try:
            abf.setSweep(sweepNumber)
        except:
            plt.plot(abf.sweepX, abf.sweepY, alpha=.5)
        
            plt.margins(0, .1)
            plt.legend()
            plt.ylabel(abf.sweepLabelY)
            plt.xlabel(abf.sweepLabelX)
            plt.title(abf.abfID)
            plt.tight_layout()
            break
    plt.show()
```

```python
for file in files:
    plot_abf(file)
```
